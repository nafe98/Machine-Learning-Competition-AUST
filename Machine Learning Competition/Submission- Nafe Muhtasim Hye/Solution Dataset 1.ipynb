# Importing Libraries

# PROCESSING DATE DATA

```python
df["Date"] = pd.to_datetime(df["Date"] , format = "%m-%d-%y")
```

```python
df["Date"].dt.day_name()
```

```python
df ["Day"] = df["Date"].dt.day_name()
```

```python
df
```

```python
df.sort_values(by = "Date", ascending = True, inplace = True)
```

# Exploratory Data Analysis(EDA)  and Correlations finding

```python
df.corr()
```

```python
import seaborn as sn
plt.figure(figsize=(15, 15))
sn.heatmap(df.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.title('', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(df["Date"], df["Daily Total Reach"], 'b.-')


plt.xlabel('Date')
plt.ylabel('Daily Total Reach')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.title('', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(df["Date"], df["Daily Page Engaged Users"], 'r.-')


plt.xlabel('Date')
plt.ylabel('Daily Page Engaged Users')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.title('', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(df["Date"], df["Lifetime Total Likes"], 'r.-')


plt.xlabel('Date')
plt.ylabel('Lifetime Total Likess')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.title('', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(df["Day"], df["Lifetime Total Likes"], 'r.-')


plt.xlabel('Day')
plt.ylabel('Lifetime Total Likess')

plt.legend()

plt.show()
```
